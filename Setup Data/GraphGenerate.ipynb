```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
methods = ['backend_search_V1', 'backend_search_V2', 'backend_search_V3']
Average_Results_Quality = [0.4567, 0.4368, 0.1152]
colors = ['skyblue', 'lightgreen', 'salmon']

# Set the style
sns.set_style("whitegrid")

# Create a figure and axis objects
plt.figure(figsize=(10, 6))

# Plot the bar chart
bars = plt.bar(methods, Average_Results_Quality, color=colors, edgecolor='black', linewidth=1)

# Add data labels on top of the bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2 , bar.get_height() + 0.01,
             '{:.4f}'.format(bar.get_height()), ha='center', color='black')

# Add title and labels
plt.title("Average Results Quality", fontsize=16)
plt.xlabel('Search Function', fontsize=14, labelpad=15)  # Adjust labelpad as needed

# Adjust y-axis limits
plt.ylim(0, 0.6)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
time = [3.555, 4.691, 5.688]
Results_Quality = [0.4567, 0.4368, 0.1152]
methods = ['Backend_Search_V1', 'Backend_Search_V2', 'Backend_Search_V3']
colors = ['skyblue', 'lightgreen', 'salmon']

# Set the style
sns.set_style("whitegrid")

# Create a figure and axis objects
plt.figure(figsize=(10, 6))

# Plot the bar chart
bars = plt.bar(time, Results_Quality, color=colors, edgecolor='black', linewidth=1)

# Add method labels on top of the bars
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width() / 2 , bar.get_height() + 0.01,
             methods[i], ha='center', color='black')

# Add title and labels
plt.title("Average Results Quality Over Time", fontsize=16)
plt.xlabel('Time', fontsize=14, labelpad=15)
plt.ylabel('Average Results Quality', fontsize=14)

# Adjust y-axis limits
plt.ylim(0, 0.6)

# Show plot
plt.tight_layout()
plt.show()

```
